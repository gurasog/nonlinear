```python
import sdeint #import itoint
import numpy as np
import matplotlib.pyplot as plt
```

# itoint(f, G, y0, tspan)  for Ito equation dy = f(y,t)dt + G(y,t)dW

```python
def f(x,t):
    return -x**(1/2)+2*(x**(1/2))*(1+x)
```

```python
def G(x,t):
    return (1+(1+x)**2)**(1/2)
```

```python
tspan = np.linspace(0.0, 2.0, 1000)
x0 = 0
```

```python
arrays=[]
d=[]
while len(arrays)<20:
    while len(d)<250:
        tspan = np.linspace(0.0, 2.0, 1000)
        x0 = 0
        res=sdeint.itoint(f, G, x0, tspan)
        d=res[~np.isnan(res)]
    arrays.append(d[0:250])
    d=[]
```

```python
arrays_for_plot=arrays
```

```python
values=[]
p=0
while len(values)<2000:
    p=p+1
    while len(d)<250:
        tspan = np.linspace(0.0, 2.0, 1000)
        x0 = 0
        res=sdeint.itoint(f, G, x0, tspan)
        d=res[~np.isnan(res)]
    values.append(d[110:111])
    print(p)
    d=[]
```
